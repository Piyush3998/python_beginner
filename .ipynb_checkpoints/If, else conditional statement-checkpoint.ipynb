```python
num = float(input('Enter any number: '))

if num % 2 == 0 and num % 5 == 0:
    print('good job')
elif num % 2 == 0 or num % 5 == 0:
    print('only one')
else:
    print('none')
```

```python
amount=0
used_units=float(input('Enter your unit value : '))

if used_units <=100:
    amount=0
    
elif used_units > 100 and used_units <= 200:
    amount=(used_units-100)*5
    
elif used_units >200:
    amount=(used_units-200)*10+500
    
print('Total amount is',amount)
```

```python
length = int(input('Enter the length of rect : '))
breadth = int(input('Enter the breadth of rect : '))

if length==breadth :
    print('This one is a SQUARE!')
else: 
    print('This one is a RECTANGLE!')
```

```python
marks = float(input('Enter marks : '))

if marks < 25 :
    print("F")
elif marks < 44 and marks > 25 :
    print("E")
elif marks < 49 and marks > 45 :
    print("D")
elif marks < 59 and marks > 50 :
    print("C")
elif marks < 79 and marks > 60 :
    print("B")
elif marks < 89 and marks > 80 :
    print("A")
elif marks <= 100 and marks > 90 :
    print("A+")
else:
    print("Error!")
    
```
