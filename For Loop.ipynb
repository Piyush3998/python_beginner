```python
#EX-2
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def sum_primes(start, end):
    prime_sum = 0
    for num in range(start, end + 1):
        if is_prime(num):
            prime_sum += num
    return prime_sum

start_range = 10
end_range = 1000

total_sum = sum_primes(start_range, end_range)
print("The sum of prime numbers between", start_range, "and", end_range, "is:", total_sum)
```

```python
#Simplified EX-2
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

start_range = 10
end_range = 1000
prime_sum = 0

for num in range(start_range, end_range + 1):
    if is_prime(num):
        prime_sum += num

print("The sum of prime numbers between", start_range, "and", end_range, "is:", prime_sum)
```
